```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

Makes sure to call encoder.fit on the Label column

```python
print('Worked')
df = pd.read_csv("tuesday_plus_cleaned.csv", skipinitialspace=True, encoding = 'latin-1')
encoder = LabelEncoder()
df['Label'] = encoder.fit_transform(df['Label'])  
df.head()
```

```python
X = df.iloc[:, 0:85]
y = df.iloc[:, 86]
```

```python
from sklearn.model_selection import train_test_split
y_encoded = encoder.fit_transform(df['Label'])

X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, random_state=42, stratify=y_encoded, test_size = 0.2)
```

```python
print(X_train.shape)
y_train.shape
X_train_full = X_train.copy()
y_train_full = y_train.copy()
```

```python
X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
dtc.get_params()
```

```python
X_train.dtypes
```

```python
X_train = X_train.select_dtypes(include=[np.number])
X_test = X_test.select_dtypes(include=[np.number])
X_train.dtypes
```

```python
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, y_pred))
print(encoder.classes_)

print("All known classes:", encoder.classes_)
print("Classes in y_test:", np.unique(y_test))
print("Classes predicted:", np.unique(y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
